```python
#Karar Yapıları ve Boole Lojiği

a = 6
b = 7

print(not a == 7 and b == 7) #True
print(a == 7 or b == 7) #True
print(a == 7 or b == 6) #False
print(not(a == 7 and b == 6)) #True
print(not a == 7 and b == 6) #False
```

```python
print('a > b :', a > b)
print('a < b :', a < b)
print('a == b :', a == b)
print('a != b :', a != b)
print('a >= b :', a >= b)
print('a <= b :', a <= b)
```
